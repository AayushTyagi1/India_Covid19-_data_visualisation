```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv("state_wise.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[1:]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Active',               
'Last_Updated_Time',    
'Migrated_Other',       
'State_code',           
'Delta_Confirmed',     
'Delta_Recovered' ,     
'Delta_Deaths','State_Notes'],inplace=True )
```

```python
df.head()
```

```python
plt.figure(1)
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(90)
plt.bar(df['State'],df['Confirmed'],width=0.9,bottom=0.25)
plt.xlabel("state")
plt.ylabel("No_of_people")
plt.legend(['Confirmed covid positive'])
plt.show()
```

```python
plt.figure(2)
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(90)
plt.bar(df['State'],df['Deaths'],width=0.9,bottom=0.25,color='r')
plt.xlabel("state")
plt.ylabel("No_of_people")
plt.legend(['Deaths'])
```

```python
plt.figure(3)
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(90)
plt.plot(df['State'],df['Confirmed'],'-o')
plt.plot(df['State'],df['Deaths'],'-o')
plt.xlabel("state")
plt.ylabel("No_of_people")
plt.legend(['Confirmed','Deaths'])
```

```python
plt.figure(5)
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(90)
xvals = df['State']
plt.bar(xvals, df['Deaths'], width = 0.8, color='r')
plt.bar(xvals, df['Confirmed'], width = 0.8, bottom=df['Deaths'], color='b')
plt.xlabel("state")
plt.ylabel("No_of_people")
plt.legend(['Death','Confirmed'])
```
